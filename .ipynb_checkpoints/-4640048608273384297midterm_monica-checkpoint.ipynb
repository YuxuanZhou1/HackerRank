### 3. Hierarchical Model

In this section, we write a model for $\textbf{describing the rate of non-violent crimes committed in each of the communities}$ in the data set.

-->nonViolPerPop

1. $\textbf{Non-herarchical model}$. Begin with a pooled model. Inspection of the histogram from the exploratory section suggests that a log-normal likelihood might fit well, so let's try the following model:


$$
\begin{align}
y_j &\sim \mathrm{Lognormal}(\mu, σ) \\
\mu &\sim \mathrm{Normal}(0, 5) \\
σ &\sim \mathrm{HalfCauchy}(1)
\end{align}
$$

Specify the model, $\textbf{sample a trace}$ , and $\textbf{draw a posterior predictive sample}$ by using $\textbf{pm.sample_postreior_predictive}$.

```python
with pm.Model() as pooled_model:
    
    # Prior
    μ = pm.Normal('μ', mu = 0, sigma = 5)
    σ = pm.HalfCauchy('σ', 1)

    # Likelihood
    y_obs = pm.Lognormal('y_obs', mu = μ, sigma = σ, observed = mid.nonViolPerPop)
    
    # Infrerence
    trace = pm.sample(2000, target_accept = 0.9, tune = 1000)
    
    replicates = pm.sample_posterior_predictive(trace)   
```

3. $\textbf{Hierarchical model}$: The posterior predictive check should indicate a reasonable fit to data; however, one might be interested in a comparison of states to one another, which the pooled model doesnt give us. So, let's set up and fit the following hierarchical normal/lognormal model:

```python
with pm.Model() as hierarchical_model:
    
    # Prior
    μ = pm.Normal('μ', mu = 0, sigma = 1)
    τ = pm.HalfCauchy('τ', 1)
    η = pm.Normal('η', mu = 0, sigma = 1, shape = 46)
    σ = pm.HalfCauchy('σ', 1)
    
    θ = pm.Deterministic('θ', μ + τ * η)
    
    #likelihood
    y_obs = pm.Lognormal('y_obs', mu = θ[mid.statecode], sigma = σ, observed = mid.nonViolPerPop)
    
    #inference
    trace_hier = pm.sample(2000, target_accept = 0.9, tune = 1000)
```

Specify the model and sample a trace. Use a forest plot or summary to examine the distributions of θ for the different states. Which state appears to $\textbf{have the lowest rate of non-violent crime}$?

```python
pm.summary(trace_hier)
```

```python
pm.forestplot(trace_hier, var_names = ['θ'])
plt.show()
```

the lowest rate of non-violent crime : $\textbf{MA(Massachusetts)}$

### 4. Linear regression


1. $\textbf{Simple linear regression}$. Fit a linear model using the above template using $\textbf{the log of PctNotHSGrad}$, to predict $\textbf{the log of nonViolPerPop}$. You may choose to standardize the log-transformed data, but it is not necessary. Setting the prior parameters μα,τα,τβ is up to you. 

```python
# make log version of outcome

nonvio_log = np.log(mid.nonViolPerPop)
grad_log = np.log(mid.PctNotHSGrad)
under_log = np.log(mid.PctPopUnderPov)
ownoc_log = np.log(mid.PctPersOwnOccup)
```

```python
μ_α = 0
τ_α = 0.3
τ_β = 0.6

with pm.Model() as linear_model:
    
    #prior
    α = pm.Normal('α', mu = μ_α, sigma =τ_α)
    β = pm.Normal('β', mu = 0, sigma = τ_β)
    σ = pm.HalfCauchy('σ', 1)  
    θ = pm.Deterministic('θ', α + β * grad_log)
    
    #likelihodd
    y_obs = pm.Normal('y_obs', mu = θ, sigma = σ, observed = nonvio_log)
    
    #inference
    trace_log = pm.sample(2000, target_accept = 0.9, tune = 1000)
```

2. $\textbf{Prior predictive check}$. Draw "10 samples" from the prior distributions you used for the linear model using $\textbf{pm.sample_prior_predictive}$  and plot the corresponding lines. $\textbf{Do the lines represent “possible” relationships?}$ If the lines are too wild, tighten your priors by adjusting τα,τβ.

```python
with linear_model:
    prior_sample = pm.sample_prior_predictive()
    posterior_sample = pm.sample_posterior_predictive(trace_log, var_names = ['α', 'β', 'y_obs'])
```

```python
# Draw 10 samples from the prior distributions 

plt.figure(figsize = (10, 7))

prior_sample_α = np.random.choice(prior_sample['α'], size = 10, replace = False)
prior_sample_β = np.random.choice(prior_sample['β'], size = 10, replace = False)


x = np.linspace(-5, 5, 1000)

for a, b in zip(prior_sample_α, prior_sample_β):
    y = a + b * x
    plt.plot(x, y, color = 'red', alpha = 0.3)

plt.show()
```

Setting alpha = 0.3, we can get a pretty tighten line plot for the result. However, it seems that the plot doesnt represent possible relationships.

4. $\textbf{Multiple regression}$. Fit a multiple linear regression by adding two more slope parameters to the model, and using PctPopUnderPov and PctPersOwnOccup as additional predictors. In this expanded model, what happens to the slope for PctNotHSGrad?

```python
μ_α = 0
τ_α = 0.3
τ_β = 0.6

with pm.Model() as linear_model1:
 
    #prior
    α = pm.Normal('α', mu = μ_α, sigma =τ_α)
    β = pm.Normal('β', mu = 0, sigma = τ_β)
    βu = pm.Normal('βu', mu= 0, sigma = τ_β)
    βo = pm.Normal('βo', mu = 0, sigma = τ_β)

    σ = pm.HalfCauchy('σ', 1)  
    θ = pm.Deterministic('θ', α + β * grad_log + βu * under_log + βo * ownoc_log)  
    
    #likelihodd
    y_obs = pm.Normal('y_obs', mu = θ, sigma = σ, observed = nonvio_log)
    
    #inference
    trace_mult = pm.sample(2000, target_accept = 0.9, tune = 1000)
```
